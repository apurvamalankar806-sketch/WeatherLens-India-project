```python
# If rain_sum > 0 → it rained (1)
# If rain_sum == 0 → no rain (0)
df['rain_tomorrow'] = (df['rain_sum'] > 0).astype(int)
print(df['rain_tomorrow'].value_counts())
```

```python
print(df['city'].unique())
```

```python
print(df['city'].value_counts())
```

```python
weather_map = {
    0: 'Clear', 1: 'Mainly Clear', 
    2: 'Partly Cloudy',
    3: 'Overcast',
    51: 'Light Drizzle', 
    53: 'Moderate Drizzle',
    55: 'Dense Drizzle', 
    61: 'Slight Rain',
    63: 'Moderate Rain',
    65: 'Heavy Rain'
}

df['weather_condition'] = df['weather_code'].map(weather_map)
print(df['weather_condition'].value_counts())
```

```python
df.columns.tolist()
```

```python
df.drop(columns=['precipitation_sum'], inplace=True)
```

```python
df.columns.tolist()
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'], dayfirst=True)
```

```python
print(df['date'].head(5))
print(df['date'].dtype)
```

```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df.columns.tolist()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Summer'
    elif month in [6, 7, 8, 9]:
        return 'Monsoon'
    elif month == 10:
        return 'Retreating Monsoon'
    elif month == 11:
        return 'Pre-Winter'

df['season'] = df['month'].apply(get_season)
print(df['season'].value_counts())
```

```python
df['temp_range'] = df['temperature_2m_max'] - df['temperature_2m_min']
```
